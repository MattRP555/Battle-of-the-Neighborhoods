Utilize the functions from practice exercises that extract category and that extract the other main pieces of information from the json that is produced. Help save some time.  

```python
def getNearbyVenues(Postcode,Borough, Neighborhoods, latitudes, longitudes):
    
    venues_list=[]
    for code, bor, neigh, lat, lng in zip(Postcode,Borough,Neighborhoods, latitudes, longitudes):
            
        # create the API request URL
        url='https://api.foursquare.com/v2/venues/explore'
        coordinates='{}, {}'.format(lat, long)
        params=dict(
                client_id = '54J2Y1NQAHOZ4KNU2YK4IWT3GUB40X3FN3NHWAQKPDWIWAW2', 
                client_secret = 'OT3ODZ1S4RJPV104FIWVNFSWRO2AB51JZNT20VU1TU5KY2O4', 
                v = '20200208',
                ll=coordinates,
                limit=20
        )
        # make the GET request
        results = requests.get(url=url,params=params).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            code, 
            lat, 
            lng,
            bor,
            neigh,
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Postal Code', 
                  'Postal Code Latitude', 
                  'Postal Code Longitude',
                  'Borough',
                  'Neighborhoods',
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

Now we are moving on the clustering of the neighborhoods. We will run K-means clusters. But first we must convert categories to dummy variables

```python
TorVenuesDummy=pd.get_dummies(TorVenues[['Venue Category']])
TorVenuesDummy['Neighborhoods']=TorVenues[['Neighborhoods']]
TorVenuesDummy['Postal Code']=TorVenues[['Postal Code']]
TorVenuesDummy['Borough']=TorVenues[['Borough']]
TorOneHot=TorVenuesDummy.groupby('Postal Code').mean().reset_index().drop(['Postal Code'],1)
TorOneHot.head()
```

```python
from sklearn.cluster import KMeans
%matplotlib inline
scores=[]
# set number of clusters
for kclusters in range(1,15):
    # run k-means clustering
    kmeans = KMeans(n_clusters=kclusters, random_state=0,n_init=15).fit(TorOneHot)

    # check cluster labels generated for each row in the dataframe
    scores.append(kmeans.score(TorOneHot)*-1)
k=np.array(range(1,15))
plt.plot(k,scores)
plt.xlabel('Number of Clusters')
plt.ylabel('scores (minimization function output)')
```

From the plot there is no clear and obvious elbow shape to it. But it does look like 6 to 8 appears to be a good choice for the clusters. In this case we will use 7 clusters. 

```python
#repeat the same calculations for kmeans but for 7 clusters. Further analyses will be continued from there.
NumClusters=7
kmeans2=KMeans(n_clusters=NumClusters, random_state=0,n_init=15).fit(TorOneHot)
print(kmeans2.labels_)
print(len(kmeans2.labels_))
```

```python
#add cluster label to TorNeigh_clean
TorNeigh_clean['Cluster']=kmeans2.labels_
TorNeigh_clean.head()
```

```python
#graphing the postal codes but adding cluster colors to it. 
mapOfClusters=folium.Map(location=[TorLatitude,TorLongitude])

colors=['red','blue','green','purple','orange','beige','pink']
for lat, long, neigh, clus in zip(TorNeigh_clean['Latitude'],TorNeigh_clean['Longitude'],TorNeigh_clean['Neighbourhood'],TorNeigh_clean['Cluster']):
    label='Neighbourhoods: {}'.format(neigh)
    label=folium.Popup(label,parse_html=True)
    folium.Marker(
        location=[lat,long],
        popup=label,
        icon=folium.Icon(color=colors[clus],icon='home')
    ).add_to(mapOfClusters)

mapOfClusters
```
